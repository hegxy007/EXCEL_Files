```python
sns.regplot(x='Sales',y='Profit',data=store,scatter_kws={"color":"blue"},line_kws={"color":"black"})
```

```python
plt.rcParams['figure.figsize']= (10,8)
```

```python
correlation_matrix= store.corr(method='pearson')
sns.heatmap(correlation_matrix,annot= True)
plt.title('Correlation Matrix for numeric features')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
store= store.drop(['Postal Code'],axis= 1)
store.head()
```

```python
store.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize=(5,5))
statewise = store.groupby(['Sub-Category'])['Profit'].sum().nlargest(50)
statewise.plot.barh()
```
